# Stock Market Forcasting using Linear Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
df
```

```python
%matplotlib inline
plt.figure(figsize=(5,5))
plt.plot(df['Month'], df['Stock Price ($)'], marker='o')
plt.legend(["Actual data"])
plt.xlabel("Months")
plt.ylabel("Stock Price ($)")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
reg = LinearRegression()
```

```python
X = np.array(df["Month"]).reshape(-1,1)
y = np.array(df["Stock Price ($)"])
```

```python
model = reg.fit(X,y)
model.coef_, model.intercept_
```

```python
predictions = model.predict(X)
```

# Errors

Mean Square Error & Mean Absolute Errors

```python
mse = mean_squared_error(predictions, df["Month"])
mae = mean_absolute_error(predictions, df["Month"])
mse, mae
```

```python
%matplotlib inline
plt.figure(figsize=(5,5))
plt.plot(df['Month'], df['Stock Price ($)'], marker='o')
plt.plot(df["Month"], predictions, marker='+')
plt.legend(["Actual data","Line of best fit"])
plt.xlabel("Months")
plt.ylabel("Stock Price ($)")
```

From the graph we can observe trend patterns in the data

## Linear Regression Model Forcast for next 6 months

```python
x_prediction = np.array([25,26,27,28,29,30]).reshape(-1,1)
y_predicted = model.predict(x_prediction)
y_predicted
```

## Final plot with the forcasted values

```python
%matplotlib inline
plt.figure(figsize=(5,5))
plt.plot(df['Month'], df['Stock Price ($)'], marker='o')
plt.plot(df["Month"], predictions, marker='+')
plt.plot(x_prediction, y_predicted, marker='*')
plt.legend(["Actual data","Line of best fit","Predicted Values"])
plt.xlabel("Months")
plt.ylabel("Stock Price ($)")
```
